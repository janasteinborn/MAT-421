## 20.1 Numerical Differentiation Problem Statement

Numerical differentiation is the process of estimating the derivative of a function using discrete data points. This method is particularly useful when analytical differentiation is complex or when working with real-world data that is sampled at discrete intervals.

We will explore finite difference approximations and higher-order derivatives, and also investigate numerical differentiation with noisy data.

## 20.2 Finite Difference Approximating Derivatives

```python
import numpy as np
import matplotlib.pyplot as plt

def forward_difference(f, x, h):
    return (f(x + h) - f(x)) / h

def backward_difference(f, x, h):
    return (f(x) - f(x - h)) / h

def central_difference(f, x, h):
    return (f(x + h) - f(x - h)) / (2 * h)

# Example function: f(x) = e^x
def f(x):
    return np.exp(x)

x = 1.0  # Point of interest
h = 0.01  # Step size

forward = forward_difference(f, x, h)
backward = backward_difference(f, x, h)
central = central_difference(f, x, h)

print(f'Forward Difference Approximation: {forward}')
print(f'Backward Difference Approximation: {backward}')
print(f'Central Difference Approximation: {central}')
```

## 20.3 Approximating of Higher Order Derivatives

```python
def second_derivative(f, x, h):
    return (f(x + h) - 2 * f(x) + f(x - h)) / (h**2)

# Compute second derivative approximation
second_deriv = second_derivative(f, x, h)
print(f'Second Derivative Approximation: {second_deriv}')
```

## 20.4 Numerical Differentiation with Noise

```python
# Simulating a noisy dataset
def noisy_function(x):
    return np.sin(x) + np.random.normal(0, 0.1, size=len(x))

x_vals = np.linspace(0, 2*np.pi, 100)
y_vals = noisy_function(x_vals)
h = x_vals[1] - x_vals[0]

# Apply central difference method to noisy data
noisy_derivative = central_difference(lambda x: np.sin(x) + np.random.normal(0, 0.1), x_vals, h)

# Plot results
plt.figure(figsize=(8,5))
plt.plot(x_vals, np.cos(x_vals), label='True Derivative (cos(x))', linewidth=2)
plt.plot(x_vals, noisy_derivative, '--', label='Numerical Derivative with Noise')
plt.legend()
plt.xlabel('x')
plt.ylabel('Derivative')
plt.title('Numerical Differentiation on Noisy Data')
plt.show()
```
