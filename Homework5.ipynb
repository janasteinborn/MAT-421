## Linear Algebra Concepts

### Linear Space
A linear space (or vector space) is a collection of vectors that can be added together and multiplied by scalars. This example demonstrates a linear combination of two vectors.

```python
import numpy as np
a = np.array([1, 2])
b = np.array([3, 4])
c = 3 * a + 2 * b
print(f'Vector a: {a}')
print(f'Vector b: {b}')
print(f'Linear combination (3*a + 2*b): {c}')
```

### Orthogonality
Two vectors are orthogonal if their dot product is zero. Here, we demonstrate orthogonality using standard basis vectors.

```python
v1 = np.array([1, 0])
v2 = np.array([0, 1])
dot_product = np.dot(v1, v2)
print(f'Vector v1: {v1}')
print(f'Vector v2: {v2}')
print(f'Dot product of v1 and v2: {dot_product}')
if dot_product == 0:
    print('The vectors are orthogonal.')
else:
    print('The vectors are not orthogonal.')
```

### Eigenvalues and Eigenvectors
Eigenvalues and eigenvectors describe how a matrix transforms space. An eigenvector remains in the same direction after transformation by the matrix, scaled by its corresponding eigenvalue.

```python
A = np.array([[4, -2], [1, 1]])
eigenvalues, eigenvectors = np.linalg.eig(A)
print(f'Matrix A:\n{A}')
print(f'Eigenvalues: {eigenvalues}')
print(f'Eigenvectors:\n{eigenvectors}')
```

## Linear Regression

### Generating Example Data
We create synthetic data that follows a linear trend with some noise.

```python
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
np.random.seed(42)
X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.randn(100, 1)
plt.figure(figsize=(8, 6))
plt.scatter(X, y, color='blue', label='Data points')
plt.xlabel('X')
plt.ylabel('y')
plt.title('Scatter Plot of Data')
plt.legend()
plt.show()
```

### Linear Regression Model
We fit a linear regression model to the data to determine the best-fit line.

```python
lin_reg = LinearRegression()
lin_reg.fit(X, y)
print(f'Intercept: {lin_reg.intercept_[0]}')
print(f'Slope: {lin_reg.coef_[0][0]}')
```

### Plotting the Regression Line
We visualize the regression line along with the original data points.

```python
X_new = np.array([[0], [2]])
y_predict = lin_reg.predict(X_new)
plt.figure(figsize=(8, 6))
plt.plot(X_new, y_predict, color='red', label='Regression line')
plt.scatter(X, y, color='blue', label='Data points')
plt.xlabel('X')
plt.ylabel('y')
plt.title('Linear Regression')
plt.legend()
plt.show()
```

The regression line effectively models the linear trend in the data.
