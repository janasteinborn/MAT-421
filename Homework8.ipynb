## 21.1 - Numerical Integration Problem Statement
Numerical integration is a powerful tool used to approximate the integral of a function \( f(x) \) over an interval \([a, b]\).
In many cases, direct analytical integration can be complex or even impossible. Thus, numerical methods provide a practical solution.

**Key Concepts:**
- **Partitioning the Interval**: The integration range \([a, b]\) is divided into smaller subintervals.
- **Summing Areas**: Approximations rely on summing areas of geometric shapes like rectangles or trapezoids.
- **Error Analysis**: The accuracy of the approximation depends on the number of subintervals and the chosen method.
- **Applications**: Numerical integration is widely used in physics, engineering, finance, and machine learning.


## 21.2 - Riemannâ€™s Integral
The Riemann integral approximates the integral of a function using rectangles.
We compute three types:
- **Left Riemann Sum**: Uses left endpoints of subintervals.
- **Right Riemann Sum**: Uses right endpoints of subintervals.
- **Midpoint Riemann Sum**: Uses midpoints of subintervals.

Demonstrating this using the function \( f(x) = x^2 \) on the interval [0,2]:

```python
import numpy as np
import matplotlib.pyplot as plt

# Function definition
f = lambda x: x**2  # Example function f(x) = x^2

# Integration limits
a, b = 0, 2
n = 10  # Number of subintervals

# Subinterval points
x = np.linspace(a, b, n+1)
dx = (b - a) / n

# Compute Riemann Sums
left_riemann = sum(f(x[:-1]) * dx)
right_riemann = sum(f(x[1:]) * dx)
midpoint_riemann = sum(f((x[:-1] + x[1:]) / 2) * dx)

# Display results
left_riemann, right_riemann, midpoint_riemann
```

```python
# Visualization of Riemann Sums
x_fine = np.linspace(a, b, 1000)
y_fine = f(x_fine)

fig, ax = plt.subplots(1, 3, figsize=(15, 5))

for i, (sum_type, title) in enumerate([(x[:-1], "Left Riemann Sum"), (x[1:], "Right Riemann Sum"), ((x[:-1] + x[1:]) / 2, "Midpoint Riemann Sum")]):
    ax[i].plot(x_fine, y_fine, 'r', label='f(x) = x^2')
    ax[i].bar(sum_type, f(sum_type), width=dx, alpha=0.5, edgecolor='black', align='edge')
    ax[i].set_title(title)
    ax[i].legend()

plt.show()
```

## 21.3 - Trapezoid Rule
The **Trapezoid Rule** approximates the integral by summing areas of trapezoids.
It provides a better approximation compared to basic Riemann sums by averaging the left and right sum contributions.

We will demonstrate this using the function \( f(x) = \sin(x) \) on the interval [0, Ï€].

```python
# Function definition
f = lambda x: np.sin(x)  # Example function f(x) = sin(x)

# Integration limits
a, b = 0, np.pi
n = 10  # Number of subintervals

# Subinterval points
x = np.linspace(a, b, n+1)
y = f(x)
dx = (b - a) / n

# Compute Trapezoidal Integral
trapezoidal_integral = dx * (sum(y[:-1] + y[1:]) / 2)

# Display result
trapezoidal_integral
```

```python
# Visualization of the Trapezoidal Rule
x_fine = np.linspace(a, b, 1000)
y_fine = f(x_fine)

plt.figure(figsize=(7, 5))
plt.plot(x_fine, y_fine, 'r', label='f(x) = sin(x)')
for i in range(n):
    plt.fill_between([x[i], x[i+1]], [y[i], y[i+1]], alpha=0.5, edgecolor='black')
plt.title("Trapezoidal Rule Approximation")
plt.legend()
plt.show()
```
